```python
import spacy
import pickle
```

```python
nlp=spacy.load('en_core_web_sm')
nlp.pipe_names
```

```python
f=open('dataset_bin','rb')
train=pickle.load(f)
f.close()
```

```python
nlp.pipe_names
```

```python
ner=nlp.get_pipe("ner")
```

```python
for _, annotations in train:
    for ent in annotations.get("entities"):
        ner.add_label(ent[2])
```

```python
disable_pipes=[pipe for pipe in nlp.pipe_names if pipe !='ner']
```

```python
import random
from spacy.util import minibatch, compounding
from spacy.training.example import Example
from pathlib import Path

with nlp.disable_pipes(*disable_pipes):
    optimizer=nlp.resume_training()

    for iteration in range(100):
        random.shuffle(train)
        losses={}
        
        batches=minibatch(train,size=compounding(1.0,4.0,1.001))
        for batch in batches:
            l=[]
            for text,entity in batch:
                doc = nlp.make_doc(text)
#                 tags=spacy.training.offsets_to_biluo_tags(nlp.make_doc(text), [entity['entities'][0]])
#                 print(tags)
                example = Example.from_dict(doc, entity)
                l.append(example)
#             print(batch)
#             text,annotation=zip(*batch)
#             print(text)
#             doc = nlp.make_doc(text)
#             example = Example.from_dict(doc, annotations)
            nlp.update(
                l,
                drop=0.5,
                losses=losses,
                sgd=optimizer
            )
            print("losses",losses)
```

```python
nlp.to_disk('./custom_nlp_for_dance')
```

```python
from spacy import displacy
q_list=[
    "Retrieve an instance which boy performing nagabandha",
    "Show a pic in which dancer is performing karkata with shanta rasa",
    "Get an image which she performing ardhachandra on left hand with rudra rasa",
    "Show me a picture which man performing mrigashirsha on left hand and kartari on right hand with hasya rasa",
    "Fetch me a picture of he performing mayura on left hand and suchi on right hand with rudra rasa",
    "Get me where female performing chakra with veera rasa",
    "Retrieve an instance of boy performing swastika with bhaya rasa",
    "Retrieve an image which woman performing mrigashirsha on left hand and mukula on right hand with bhaya rasa",
    "Dancer performing katakamudra",
]
for txt in q_list:
    doc=nlp(txt)
    displacy.render(nlp(doc.text),style='ent',jupyter=True)
```
